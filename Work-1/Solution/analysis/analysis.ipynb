```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('time_results/results.csv')

plt.figure(figsize=(10, 6))
plt.plot(df['Size'], df['BinarySearch'], marker='o', label='Binary Search')
plt.plot(df['Size'], df['MergeSort'], marker='s', label='Merge Sort')
plt.plot(df['Size'], df['QuickSort'], marker='^', label='Quick Sort')

plt.xlabel('Array Size')
plt.ylabel('Execution Time (microseconds)')
plt.title('Algorithm Performance Comparison')
plt.legend()
plt.grid(True, alpha=0.3)
plt.xticks(df['Size'])

plt.savefig('time_results/performance_chart.png', dpi=150, bbox_inches='tight')
plt.show()
```
